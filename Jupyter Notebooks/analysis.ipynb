NOW IT'S TIME TO CLEAN THIS ALL UP

Person Alter Name - Simple lookup table. We strip whitespace from names, drop nulls (a name with no value is useless), drop empty strings, and remove duplicates on the (person_mal_id, alt_name) pair since the same alias listed twice for the same person is redundant.

```python
# =============================================================================
# ANIME DATASET — DATA CLEANING
# File: person_alternate_names.csv
# =============================================================================
import os

BASE_PATH = '/Users/alexandrubudau/Desktop/HCI-PROJECT/datasets/'
OUT_PATH  = '/Users/alexandrubudau/Desktop/HCI-PROJECT/cleaned datasets/'
os.makedirs(OUT_PATH, exist_ok=True)

print("=" * 60)
print("CLEANING: person_alternate_names.csv")
print("=" * 60)

df = pd.read_csv(BASE_PATH + 'person_alternate_names.csv')
raw = df.copy()

print(f"\nRaw shape: {df.shape}")
print(df.head(5))

# --- Strip whitespace ---
df['alt_name'] = df['alt_name'].str.strip()

# --- Drop nulls ---
df = df.dropna(subset=['alt_name'])

# --- Drop empty strings ---
df = df[df['alt_name'] != '']

# --- Drop duplicates ---
df = df.drop_duplicates(subset=['person_mal_id', 'alt_name'])

# --- Summary ---
print(f"\n{'─'*60}")
print(f"  Rows before : {len(raw):,}")
print(f"  Nulls removed       : {raw['alt_name'].isnull().sum()}")
print(f"  Duplicates removed  : {raw.duplicated(subset=['person_mal_id','alt_name']).sum()}")
print(f"  Rows after  : {len(df):,}")
print(f"{'─'*60}")

df.to_csv(OUT_PATH + 'person_alternate_names_clean.csv', index=False)
print("Saved → person_alternate_names_clean.csv")
```

Characters — We drop the url column since links aren't useful for data analysis. name is critical so rows missing it are dropped. about and favorites are optional fields so instead of dropping we fill them with '' and 0 respectively. Dedup on character_mal_id since each character should appear once.

```python
# =============================================================================
# ANIME DATASET — DATA CLEANING
# File: characters.csv
# =============================================================================

import pandas as pd
import os

BASE_PATH = '/Users/alexandrubudau/Desktop/HCI-PROJECT/datasets/'
OUT_PATH  = '/Users/alexandrubudau/Desktop/HCI-PROJECT/cleaned datasets/'

print("=" * 60)
print("CLEANING: characters.csv")
print("=" * 60)

df = pd.read_csv(BASE_PATH + 'characters.csv')
raw = df.copy()

print(f"\nRaw shape: {df.shape}")
print(df.head(5))

# --- Drop URL column (not useful for analysis) ---
df = df.drop(columns=['url'], errors='ignore')

# --- Drop rows with no name (critical field) ---
df = df.dropna(subset=['name'])

# --- Strip text fields ---
df['name']       = df['name'].str.strip()
df['name_kanji'] = df['name_kanji'].str.strip()

# --- Fill missing about with empty string ---
df['about'] = df['about'].fillna('')

# --- Fill missing favorites with 0 ---
df['favorites'] = df['favorites'].fillna(0).astype(int)

# --- Drop duplicates on ID ---
df = df.drop_duplicates(subset=['character_mal_id'])

# --- Summary ---
print(f"\n{'─'*60}")
print(f"  Rows before         : {len(raw):,}")
print(f"  Nulls (name)        : {raw['name'].isnull().sum()}")
print(f"  Duplicates removed  : {raw.duplicated(subset=['character_mal_id']).sum()}")
print(f"  Rows after          : {len(df):,}")
print(f"{'─'*60}")

df.to_csv(OUT_PATH + 'characters_clean.csv', index=False)
print("Saved → characters_clean.csv")
```

**In this part of cleaning, a score of 0 or 0 episodes means something very different from "unknown". Keeping them as NaN means pandas and seaborn will simply skip them in calculations and plots, which is the correct behaviour.**

*Details — The richest file. We drop url and image_url, parse start_date and end_date into proper datetime objects and extract start_year and start_month for trend analysis. All numeric fields are coerced to numbers. Missing score and episodes are kept as NaN intentionally — they're optional and dropping them would lose too many rows. Dedup on mal_id.*

```python
BASE_PATH = '/Users/alexandrubudau/Desktop/HCI-PROJECT/datasets/'
OUT_PATH  = '/Users/alexandrubudau/Desktop/HCI-PROJECT/cleaned datasets/'
os.makedirs(OUT_PATH, exist_ok=True)

print("=" * 60)
print("CLEANING: details.csv")
print("=" * 60)

df = pd.read_csv(BASE_PATH + 'details.csv')
raw = df.copy()

print(f"\nRaw shape: {df.shape}")
print(df.head(5))

# --- Drop URL/image columns ---
df = df.drop(columns=['url', 'image_url'], errors='ignore')

# --- Drop rows with no title (critical) ---
df = df.dropna(subset=['title'])

# --- Parse dates ---
for col in ['start_date', 'end_date']:
    df[col] = pd.to_datetime(df[col], errors='coerce', utc=True)
    df[col] = df[col].dt.tz_localize(None)

# --- Extract useful date parts ---
df['start_year']  = df['start_date'].dt.year
df['start_month'] = df['start_date'].dt.month

# --- Numeric fields ---
df['score']    = pd.to_numeric(df['score'],    errors='coerce')
df['episodes'] = pd.to_numeric(df['episodes'], errors='coerce')
df['year']     = pd.to_numeric(df['year'],     errors='coerce')
df['rank']     = pd.to_numeric(df['rank'],     errors='coerce')
df['popularity']= pd.to_numeric(df['popularity'], errors='coerce')
df['members']  = pd.to_numeric(df['members'],  errors='coerce')
df['favorites']= pd.to_numeric(df['favorites'],errors='coerce')
df['scored_by']= pd.to_numeric(df['scored_by'],errors='coerce')

# --- Strip text fields ---
df['title']   = df['title'].str.strip()
df['type']    = df['type'].str.strip()
df['status']  = df['status'].str.strip()
df['source']  = df['source'].str.strip()
df['rating']  = df['rating'].str.strip()
df['season']  = df['season'].str.strip()

# --- Fill missing synopsis ---
df['synopsis'] = df['synopsis'].fillna('')

# --- Drop duplicates ---
df = df.drop_duplicates(subset=['mal_id'])

# --- Summary ---
print(f"\n{'─'*60}")
print(f"  Rows before         : {len(raw):,}")
print(f"  Nulls (title)       : {raw['title'].isnull().sum()}")
print(f"  Duplicates removed  : {raw.duplicated(subset=['mal_id']).sum()}")
print(f"  Missing scores      : {df['score'].isnull().sum():,}")
print(f"  Missing episodes    : {df['episodes'].isnull().sum():,}")
print(f"  Rows after          : {len(df):,}")
print(f"{'─'*60}")
print(f"\nType distribution:\n{df['type'].value_counts()}")
print(f"\nStatus distribution:\n{df['status'].value_counts()}")

df.to_csv(OUT_PATH + 'details_clean.csv', index=False)
print("\nSaved → details_clean.csv")
```

Ratings — By far the largest file (124M+ rows) so it's processed in chunks to avoid memory crashes. For each chunk: drop nulls on username and anime_id, validate status against a known set of valid values, validate score is between 0–10, fill missing counts with 0. After all chunks are concatenated, a final global dedup is done on (username, anime_id) since a user should only have one rating per anime.

```python
BASE_PATH = '/Users/alexandrubudau/Desktop/HCI-PROJECT/datasets/'
OUT_PATH  = '/Users/alexandrubudau/Desktop/HCI-PROJECT/cleaned datasets/'
os.makedirs(OUT_PATH, exist_ok=True)

print("=" * 60)
print("CLEANING: ratings.csv  [chunk processing]")
print("=" * 60)

CHUNK_SIZE    = 5_000_000
VALID_STATUSES = {'watching', 'completed', 'on_hold', 'dropped', 'plan_to_watch'}

chunks_clean = []
total_raw    = 0
total_clean  = 0
chunk_num    = 0

for chunk in pd.read_csv(BASE_PATH + 'ratings.csv', chunksize=CHUNK_SIZE):
    chunk_num  += 1
    total_raw  += len(chunk)

    # --- Drop nulls on critical fields ---
    chunk = chunk.dropna(subset=['username', 'anime_id'])

    # --- Normalise status ---
    chunk['status'] = chunk['status'].str.strip().str.lower()
    chunk = chunk[chunk['status'].isin(VALID_STATUSES)]

    # --- Validate score (must be 0–10 or NaN) ---
    chunk['score'] = pd.to_numeric(chunk['score'], errors='coerce')
    chunk = chunk[(chunk['score'].isna()) | (chunk['score'].between(0, 10))]

    # --- Normalise is_rewatching ---
    chunk['is_rewatching'] = chunk['is_rewatching'].fillna(0).astype(int)

    # --- Fill missing episode count ---
    chunk['num_watched_episodes'] = pd.to_numeric(
        chunk['num_watched_episodes'], errors='coerce'
    ).fillna(0).astype(int)

    # --- Drop duplicates within chunk ---
    chunk = chunk.drop_duplicates(subset=['username', 'anime_id'])

    total_clean += len(chunk)
    chunks_clean.append(chunk)

    print(f"  Chunk {chunk_num:>3} — raw: {len(chunk) + (total_raw - total_clean - len(chunk)):,} → kept: {len(chunk):,}")

print("\nConcatenating all chunks...")
df = pd.concat(chunks_clean, ignore_index=True)

# --- Final global dedup across chunks ---
before_global_dedup = len(df)
df = df.drop_duplicates(subset=['username', 'anime_id'])
print(f"  Cross-chunk duplicates removed: {before_global_dedup - len(df):,}")

# --- Summary ---
print(f"\n{'─'*60}")
print(f"  Total raw rows    : {total_raw:,}")
print(f"  Total clean rows  : {len(df):,}")
print(f"  Total removed     : {total_raw - len(df):,}")
print(f"{'─'*60}")
print(f"\nStatus distribution:\n{df['status'].value_counts()}")
print(f"\nScore distribution:\n{df['score'].value_counts().sort_index()}")

df.to_csv(OUT_PATH + 'ratings_clean.csv', index=False)
print("\nSaved → ratings_clean.csv")
```

```python
def parse_list_col(val):
    try:
        parsed = ast.literal_eval(val)
        if isinstance(parsed, list):
            return parsed
    except:
        pass
    return []

df['genres_parsed'] = df['genres'].apply(parse_list_col)

# Explode genres
df_genres = df[['decade', 'genres_parsed']].explode('genres_parsed')
df_genres = df_genres.rename(columns={'genres_parsed': 'genre'})
df_genres = df_genres[df_genres['genre'].notna() & (df_genres['genre'] != '')]

# Count per decade per genre
decade_genre = (
    df_genres.groupby(['decade', 'genre'])
    .size()
    .reset_index(name='count')
)

# Normalise as % of total anime that decade
total_per_decade = df.groupby('decade').size().reset_index(name='total')
decade_genre = decade_genre.merge(total_per_decade, on='decade')
decade_genre['pct'] = (decade_genre['count'] / decade_genre['total'] * 100).round(2)

# Top 8 genres by total count
top_genres = (
    decade_genre.groupby('genre')['count']
    .sum()
    .sort_values(ascending=False)
    .head(8)
    .index.tolist()
)

df_top = decade_genre[decade_genre['genre'].isin(top_genres)]
pivot2 = df_top.pivot(index='decade', columns='genre', values='pct').fillna(0)

# --- Plot ---
palette = [
    '#E63946', '#457B9D', '#2A9D8F', '#E9C46A',
    '#F4A261', '#A8DADC', '#6D2B8C', '#2B9348'
]

fig, ax = plt.subplots(figsize=(16, 7))

ax.stackplot(
    pivot2.index,
    [pivot2[genre] for genre in pivot2.columns],
    labels=pivot2.columns,
    colors=palette,
    alpha=0.85
)

ax.set_title('Genre Popularity Trends by Decade (% of all anime)',
             fontsize=16, fontweight='bold', pad=20)
ax.set_xlabel('Decade', fontsize=12)
ax.set_ylabel('% of Anime in that Decade', fontsize=12)
ax.legend(title='Genre', fontsize=10, bbox_to_anchor=(1.01, 1), loc='upper left')
ax.xaxis.set_major_locator(ticker.MultipleLocator(10))
ax.grid(axis='y', linestyle='--', alpha=0.3)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig(OUT_PATH + 'chart2_genre_trends_area.png', dpi=150, bbox_inches='tight')
plt.show()
```
