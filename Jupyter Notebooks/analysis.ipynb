HCI-PROJECT, BUDAU ALEXANDRU 



This project analyzes anime-related data to explore trends, popularity, and geographic distribution of anime audiences. The analysis is developed in Jupyter Notebooks, which provide an interactive environment for combining code, visualizations, and explanations. Python is used as the main programming language together with data analysis libraries such as Pandas for data manipulation and Seaborn for statistical visualization. Geographic insights are generated using GeoSeaborn and Folium to create spatial and interactive maps. The goal of the project is to transform raw anime data into clear and meaningful visual insights.
