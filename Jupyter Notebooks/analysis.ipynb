```python
df = pd.read_csv('/Users/alexandrubudau/Desktop/HCI-PROJECT/datasets/person_alternate_names.csv')

# NaN Report
nan_per_col = df.isnull().sum()
nan_pct     = (df.isnull().mean() * 100).round(2)
total_nan   = nan_per_col.sum()
total_cells = df.shape[0] * df.shape[1]

result = pd.DataFrame({
    'column'    : nan_per_col.index,
    'nan_count' : nan_per_col.values,
    'nan_%'     : nan_pct.values
})
result = result[result['nan_count'] > 0].reset_index(drop=True)

print(f"\n{'─'*50}")
print(f"  NaN REPORT — person_alternate_names")
print(f"{'─'*50}")
if result.empty:
    print("  No NaN values found.")
else:
    print(result.to_string(index=False))
print(f"\n  Total NaN     : {total_nan:,}")
print(f"  Total cells   : {total_cells:,}")
print(f"  NaN overall % : {(total_nan / total_cells * 100):.2f}%")
```

NOW IT'S TIME TO CLEAN THIS ALL UP

```python
# =============================================================================
# ANIME DATASET — DATA CLEANING
# File: person_alternate_names.csv
# =============================================================================

import pandas as pd
import os

BASE_PATH = '/Users/alexandrubudau/Desktop/HCI-PROJECT/datasets/'
OUT_PATH  = '/Users/alexandrubudau/Desktop/HCI-PROJECT/cleaned datasets/'
os.makedirs(OUT_PATH, exist_ok=True)

print("=" * 60)
print("CLEANING: person_alternate_names.csv")
print("=" * 60)

df = pd.read_csv(BASE_PATH + 'person_alternate_names.csv')
raw = df.copy()

print(f"\nRaw shape: {df.shape}")
print(df.head(5))

# --- Strip whitespace ---
df['alt_name'] = df['alt_name'].str.strip()

# --- Drop nulls ---
df = df.dropna(subset=['alt_name'])

# --- Drop empty strings ---
df = df[df['alt_name'] != '']

# --- Drop duplicates ---
df = df.drop_duplicates(subset=['person_mal_id', 'alt_name'])

# --- Summary ---
print(f"\n{'─'*60}")
print(f"  Rows before : {len(raw):,}")
print(f"  Nulls removed       : {raw['alt_name'].isnull().sum()}")
print(f"  Duplicates removed  : {raw.duplicated(subset=['person_mal_id','alt_name']).sum()}")
print(f"  Rows after  : {len(df):,}")
print(f"{'─'*60}")

df.to_csv(OUT_PATH + 'person_alternate_names_clean.csv', index=False)
print("Saved → person_alternate_names_clean.csv")
```
